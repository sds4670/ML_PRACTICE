```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline
sns.set(color_codes=True)
```

DROP COLUMNS

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

 Renaming the columns
 most of the column names are very confusing to read, so I just tweaked their column names. This is a good approach it improves the readability of the data set.

```python
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

Dropping the missing or null values

```python
print(df.isnull().sum())
```

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

```python
print(df.isnull().sum())
```

 Detecting Outliers

An outlier is a point or set of points that are different from other points. Sometimes they can be very high or very low. It's often a good idea to detect and remove the outliers. Because outliers are one of the primary reasons for resulting in a less accurate model. Hence it's a good idea to remove them.



```python
sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
numeric_df = df.select_dtypes(include=np.number)
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

Histogram

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Price'], kde=True)
plt.title("Price Distribution")
plt.show()
```

```python
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

Scatterplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

features = ['HP', 'Cylinders', 'MPG-H', 'MPG-C']

for col in features:
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=df[col], y=df['Price'])
    plt.title(f"Price vs {col}")
    plt.xlabel(col)
    plt.ylabel("Price")
    plt.show()
```

Heat Maps

```python
plt.figure(figsize=(10,5))
numeric_df = df.select_dtypes(include=np.number)
c= numeric_df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

The correlation heatmap shows strong correlation between MPG-H and MPG-C, indicating redundancy. One of these features will be removed during feature selection to avoid multicollinearity in regression-based models.
